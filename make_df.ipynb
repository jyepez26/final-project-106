```python
import pandas as pd
```

```python
start_cd = 'all_data/wearable_data/Data/'
```

```python
students = ['S'+str(i) for i in range(1,11)]
```

```python
tests = ['Final', 'Midterm 1', 'Midterm 2']
```

```python
ACC_df_prep = pd.DataFrame()
for s in students:
    for t in tests:
        df_paused = pd.read_csv(start_cd+s+'/'+t+'/ACC.csv')
        df_paused = df_paused.drop(0).reset_index().drop(columns=['index'])
        cols = df_paused.columns
        df_paused = df_paused.rename(columns={cols[0]:'x',cols[1]:'y',cols[-1]:'z'})
        df_paused['student'] = s
        df_paused['test'] = t
        ACC_df_prep = pd.concat([ACC_df_prep,df_paused])

ACC_df = ACC_df_prep
ACC_df.head()
```
